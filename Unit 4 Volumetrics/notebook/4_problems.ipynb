# Problem 4.3

Comment: this problem is more convenient to be worked on Excel spreadsheet

```
"Task. calculate OOIP using trapezoidal rule"

# known
Bo = 1.3 # in RB/STB

# write Table 4.7 into numpy arrays

# N is originally: N = h*poro*(1-Sw) / Bo, written as N = term / Bo. The given data is in term (h*poro*(1-Sw))
term = np.array([0, 5, 10, 15, 20, 25, 27.5])
area = np.array([450, 375, 303, 231, 154, 74, 0]) # in acre

# convert area in acre to ft2
area = area * 2.788E+07 # acre to ft2

# compute N = term / Bo
N = (term / Bo) * (1 / 5.61458) # result in STB/ft2

# using trapezoidal to calculate OIP
# notice that in 6th and 7th data, the space between them is different from 1st, 2nd, ..., 5th. calculate the delta_N

delta_N = [j-i for i, j in zip(N[:-1], N[1:])]
print(delta_N)
```

```
# because data is IRREGULARLY SPACED, trapezoidal rule will use its traditional form
# V = ((0.5 * deltaz1) * (a1+a2)) + ((0.5 * deltaz2) * (a2+a3)) + ... + ((0.5 * deltazn) * (an-1 + an))

# the last (7th) area is more conveniently to be deleted from the array
area = np.delete(area, -1)

oip_trapeze = [((i+j)*(k/2)) for i, j, k in zip(area[:-1], area[1:], delta_N[:-1])]
oip_trapeze = np.sum(oip_trapeze)
print("OIP of oil reservoir:", oip_trapeze / 1E+06, "million STB")
```

# Problem 4.5 (OOIP)

```
"Task. determine OOIP"

# known

Bo = 1.324 # in RB/STB
owc = -5140

# data

welldat = pd.read_csv('/content/reservoir-engineering/Unit 4 Volumetrics/data/Table 4.9-Well Data for Problem 4.5.csv')
x, y, depth = welldat.x, welldat.y, welldat.depth
z = depth
```

```
# structural map with tricontouring

# delaunay triangulation / creating triangles from data points
from delaunay import triangulation

smoothness = 5

triang, refiner, tri_refi, z_test_refi = triangulation(x, y, z, smoothness)

level = 10
label = 2

#-----------------------------------------------------------------------------
# Plot the triangulation and the high-res iso-contours
#-----------------------------------------------------------------------------

### This is a code, not a function. Is preceded by triangulation using delaunay.py function. After running it, copy this code to plot the result. 

import matplotlib.tri as tri
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# INPUTS: 
# triang, refiner, tri_refi, z_test_refi: results from delaunay.py function
# level: controls how dense the contours will be, 10 is common, higher level = denser contour
# label: controls the appearance of labeling contour value, 3 is common, higher label = more frequent the contour label will be, 1 = per contour is labeled
  
plt.figure(figsize=(20, 10))
plt.gca().set_aspect('equal')
plt.triplot(triang, lw=0.5, color='white')

levels = np.linspace(min(z)+10, max(z)-10, level)  ##contour range, divided by 10 levels
cmap = cm.get_cmap(name='terrain', lut=None)
fig = plt.tricontourf(tri_refi, z_test_refi, levels=levels, cmap=cmap)
fig = plt.tricontour(tri_refi, z_test_refi, levels=levels,
                 colors=['0.25', '0.5', '0.5', '0.5', '0.5'],
                 linewidths=[1.0, 0.5, 0.5, 0.5, 0.5])
plt.plot(x, y, 'ko', ms=3) # plot the well points
plt.clabel(fig, fig.levels[::label], inline=1, fontsize=10) # give labels for contours, fig.levels[::x] controls the frequency of labels
                                                          # lesser ::x, more populated by labels
plt.title("Structure Map using Triangulation Method", pad=10, size=20)

plt.show()
```

```
"Step 2. Compute N for each well data"

#-----------------------------------------------------------------------------
# CONVERSION !!!
#-----------------------------------------------------------------------------

# N = h*poro*(1-Sw) / Bo
# h in ft, Bo in RB/STB. N must be in STB/ft2
# N = ft * (STB/RB), RB must be converted to ft3
# 1 RB (barrel) = 5.61458 ft3
# so, N = (ft * (STB/RB)) * (1 / 5.61458) = STB/ft2

N = welldat.h * welldat.poro * (1 - welldat.sw) / Bo 
N = N * (1 / 5.61458) # conversion to STB/ft2
Ndf = pd.DataFrame({"well_id": welldat.well_id, "N (STB/ft2)": N})
Ndff = pd.merge(welldat, Ndf, on='well_id')
Ndff
```

# Problem 4.6 (OGIP)

```
"Task. determine OGIP"

# known

Bg = 0.002154 # in res m3/std m3

# data

welldat = pd.read_csv('/content/reservoir-engineering/Unit 4 Volumetrics/data/Table 4.10-Well Data for Problem 4.6.csv')
x, y, depth = welldat.x, welldat.y, welldat.depth
z = depth
```

```
# structural map with tricontouring

# delaunay triangulation / creating triangles from data points
from delaunay import triangulation

smoothness = 5

triang, refiner, tri_refi, z_test_refi = triangulation(x, y, z, smoothness)

level = 10
label = 2

#-----------------------------------------------------------------------------
# Plot the triangulation and the high-res iso-contours
#-----------------------------------------------------------------------------

### This is a code, not a function. Is preceded by triangulation using delaunay.py function. After running it, copy this code to plot the result. 

import matplotlib.tri as tri
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# INPUTS: 
# triang, refiner, tri_refi, z_test_refi: results from delaunay.py function
# level: controls how dense the contours will be, 10 is common, higher level = denser contour
# label: controls the appearance of labeling contour value, 3 is common, higher label = more frequent the contour label will be, 1 = per contour is labeled
  
plt.figure(figsize=(20, 10))
plt.gca().set_aspect('equal')
plt.triplot(triang, lw=0.5, color='white')

levels = np.linspace(min(z)+10, max(z)-10, level)  ##contour range, divided by 10 levels
cmap = cm.get_cmap(name='terrain', lut=None)
fig = plt.tricontourf(tri_refi, z_test_refi, levels=levels, cmap=cmap)
fig = plt.tricontour(tri_refi, z_test_refi, levels=levels,
                 colors=['0.25', '0.5', '0.5', '0.5', '0.5'],
                 linewidths=[1.0, 0.5, 0.5, 0.5, 0.5])
plt.plot(x, y, 'ko', ms=3) # plot the well points
plt.clabel(fig, fig.levels[::label], inline=1, fontsize=10) # give labels for contours, fig.levels[::x] controls the frequency of labels
                                                          # lesser ::x, more populated by labels
plt.title("Structure Map using Triangulation Method", pad=10, size=20)

plt.show()
```

```
"Step 2. Compute N for each well data" 

#-----------------------------------------------------------------------------
# CONVERSION !!!
#-----------------------------------------------------------------------------

# Bo = Bo * (1 / 6.28981) # m3/m3 to STB/m3

N = welldat.h * welldat.poro * (1 - welldat.sw) / Bg 

N = N * 35.3146667 # convert to scf (1 m3 = 35.31... ft3)

Ndf = pd.DataFrame({"well_id": welldat.well_id, "N (scf/m2)": N})
Ndff = pd.merge(welldat, Ndf, on='well_id')
Ndff
```

```
"Task 3. plot contour map of GIP from the resulted N data"

import scipy
from scipy.interpolate import griddata

z = N # z is now the N data

xi = np.linspace(min(x), max(x), 100)
yi = np.linspace(min(y), max(y), 100)
xi, yi = np.meshgrid(xi, yi)

# interpolate
zi = griddata((x,y),z,(xi,yi),method='cubic')

cmap = cm.get_cmap(name='terrain', lut=None)

# levels2 = np.linspace(min(z), max(z), 10)  ##contour range, divided by 10 levels
levels2 = np.arange(min(z), max(z), 5000)

plt.figure(figsize=(10, 10))
figs = plt.contour(xi, yi, zi, levels=levels2, colors=['0.25', '0.5', '0.25', '0.5', '0.25'], linewidths=[1.0, 0.5, 1, 0.5, 1])
plt.contourf(xi,yi,zi, cmap="terrain")
plt.plot(x, y, 'ko', ms=3) # plot the well points
plt.clabel(figs, figs.levels[::2], inline=1, fontsize=10) # give labels for contours, fig.levels[::x] controls the frequency of labels
                                                        # lesser ::x, more populated by labels
plt.title("GIP contour map", pad=10, size=20)
plt.colorbar()
plt.show()
```

```
"calculate area of N contours"

# Use Green's theorem to compute the area
# enclosed by the given contour.
def area(vs):
    a = 0
    x0,y0 = vs[0]
    for [x1,y1] in vs[1:]:
        dx = x1-x0
        dy = y1-y0
        a += 0.5*(y0*dx - x0*dy)
        x0 = x1
        y0 = y1
    return a

# Get one of the contours from the plot.

a_array = []

for i in range(len(levels2)):
    contour = figs.collections[i]
    vs = contour.get_paths()[0].vertices
    # Compute area enclosed by vertices.
    a = area(vs)
    a_array.append(float(np.abs(a)))
    # print("r = " + str(levels[i]) + ": a =" + str(a))

areas = pd.DataFrame({"OIP Contour(STB)": levels2, "Area(m2)": a_array})
areas
```

```
"Task 4. calculate total GIP"

# Method 1. Trapezoidal rule

from trapezoid import trapezoid

gip_trapezo = trapezoid(a_array, 5000)
print("Gas in Place using Trapezoidal Method:", gip_trapezo / 1E+09, "billion scf")
```

```
# Method 2. 1/3 Simpson

from simpson import simpson

gip_simpson = simpson(a_array, 5000)
print("Gas in Place using 1/3 Simpson's Method:", gip_simpson / 1E+09, "billion scf")
```

```
# Method 3. Pyramidal

from pyramidal import pyramid

gip_pyramidal = pyramid(a_array, 5000)
print("Gas in Place using Pyramidal Method:", gip_pyramidal / 1E+09, "billion scf")
```
